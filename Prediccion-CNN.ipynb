```python
import numpy as np
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.models import load_model

import matplotlib.pyplot as plt
```

```python
longitud, altura = 100, 100
modelo = './modeloCaballoGallina/modelo.h5'
pesos_modelo = './modeloCaballoGallina/pesos.h5'
model = load_model(modelo)
model.load_weights(pesos_modelo)
```

```python
def predict(file):
  x = load_img(file, target_size=(longitud, altura))

  plt.figure()
  plt.imshow(x)
  plt.show()
    
  x = img_to_array(x)
  x = np.expand_dims(x, axis=0)
  array = model.predict(x)
  result = array[0]
  answer = np.argmax(result)
  if answer == 0:
    print("pred: Caballo")
  elif answer == 1:
    print("pred: Gallina") 

  return answer, result, array
```

```python
predict('./Ndatos/validacion/caballo/OIP-2supbKKGV6nbGie47XAykgHaFj.jpeg')
```
